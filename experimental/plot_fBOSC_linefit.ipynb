```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import pandas as pd
from scipy import stats
from matplotlib.pyplot import figure
```

```python
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
```

```python
df = pd.read_csv('alpha_theta_knee.csv')
df.head()
```

```python
ax = sns.violinplot(x="condition", y="RMSE",
                    palette=sns.color_palette(['lightgray','lightgray','lightgray']),
                    linewidth=2.5,cut=0,inner='quartile', data=df)
#ax = sns.boxplot(x="group", y="peak_freq",palette=sns.color_palette(['lightgray','lightgray']), data=df)
ax = sns.swarmplot(x="condition", y="RMSE",
                   palette=sns.color_palette(['#ff5c04','#c8588e','#045cff']),alpha=0.7,size=1.5, data=df)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_ylabel("RMSE",fontsize=25,fontname="Arial")
ax.set_xlabel("",fontsize=25,fontname="Arial")
#ax.set_xticklabels(['Control','Visual\nSnow'])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig = ax.get_figure()
fig.show()
fig.savefig('alpha_theta_knee.png', bbox_inches='tight',dpi=600)
```
