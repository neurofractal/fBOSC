```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import pandas as pd
from scipy import stats
from matplotlib.pyplot import figure
```

```python
sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
```

```python
df = pd.read_csv('group_gamma.csv')
df.head()
```

```python
ax = sns.violinplot(x="group", y="peak_freq",palette=sns.color_palette(['silver','lightgray']),linewidth=2.5,cut=1,inner='quartile', data=df)
#ax = sns.boxplot(x="group", y="peak_freq",palette=sns.color_palette(['lightgray','lightgray']), data=df)
ax = sns.swarmplot(x="group", y="peak_freq",palette=sns.color_palette(['#ff5c04','#045cff']),alpha=0.7,size=10, data=df)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_ylabel("Peak \nFrequency (Hz)",fontsize=25,fontname="Arial")
ax.set_xlabel("",fontsize=25,fontname="Arial")
ax.set_xticklabels(['Control','Visual\nSnow'])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig = ax.get_figure()
fig.savefig('peak_freq_gamma.png', bbox_inches='tight',dpi=600)
```

```python
ax = sns.violinplot(x="group", y="pow_change",palette=sns.color_palette(['silver','lightgray']),linewidth=2.5,cut=1,inner='quartile', data=df)
#ax = sns.boxplot(x="group", y="peak_freq",palette=sns.color_palette(['lightgray','lightgray']), data=df)
ax = sns.swarmplot(x="group", y="pow_change",palette=sns.color_palette(['#ff5c04','#045cff']),alpha=0.7,size=10, data=df)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_ylabel("Power Change\n (dB)",fontsize=25,fontname="Arial")
ax.set_xlabel("",fontsize=25,fontname="Arial")
ax.set_xticklabels(['Control','Visual\nSnow'])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig = ax.get_figure()
fig.savefig('db_gamma.png', bbox_inches='tight',dpi=600)
```

```python
ax = sns.violinplot(x="group", y="baseline_pow",palette=sns.color_palette(['silver','lightgray']),linewidth=2.5,cut=1,inner='quartile', data=df)
#ax = sns.boxplot(x="group", y="peak_freq",palette=sns.color_palette(['lightgray','lightgray']), data=df)
ax = sns.swarmplot(x="group", y="baseline_pow",palette=sns.color_palette(['#ff5c04','#045cff']),alpha=0.7,size=10, data=df)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_ylabel("Baseline\n Power (T)",fontsize=25,fontname="Arial")
ax.set_xlabel("",fontsize=25,fontname="Arial")
ax.set_xticklabels(['Control','Visual\nSnow'])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.offsetText.set_fontsize(14)
fig = ax.get_figure()
fig.savefig('gamma_baseline.png', bbox_inches='tight',dpi=600)
```
